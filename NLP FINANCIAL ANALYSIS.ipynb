```python
doc=nlp(text)

from spacy import displacy
displacy.render(doc,style="ent")
```

```python
ruler=nlp.add_pipe("entity_ruler",before="ner")
```

```python
let="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
pattern=[]
for x in symbol:
    pattern.append({"label":"STOCK","pattern":x})
    for l in let:
        pattern.append({"label":"STOCK","pattern":x+f".{l}"})
for x in company_name:
    pattern.append({"label":"ORG","pattern":x})
        
ruler.add_patterns(pattern)
```

```python
doc=nlp(text)
for x in doc[:300].ents:
    print(x.text,x.label_)
```

```python
for x in exchange1:
    pattern.append({"label":"STOCK_EXCHANGE","pattern":x})
for x in exchange:
    pattern.append({"label":"STOCK_EXCHANGE","pattern":x})
    pattern.append({"label":"STOCK_EXCHANGE","pattern":"."+x})
```

```python
for x in IndexName:
    pattern.append({"label":"INDEX_NAME","pattern":x})
    word=x.split()
    pattern.append({"label":"INDEX_NAME","pattern":" ".join(word[:2])})
for x in IndexSymbol:
    pattern.append({"label":"INDEX","pattern":x})
    pattern.append({"label":"INDEX","pattern":"."+x})


for x in country:
    pattern.append({"label":"GPE","pattern":x})
    if x!="U.S.":
        pattern.append({"label":"GPE","pattern":"U.S."})
    

```

```python
ruler.add_patterns(pattern)
```
